- 6. Q.A<br>
    Q : accuracy와 interpretability 사이에 간극(tension)이 존재한다고 하였는데, 이는 무엇을 의미할까요?<br>
   A : accuracy와 interpretability 사이에 정비례 혹은 반비례처럼 특정한 해석가능한 관계까 존재하지 않음을 의미하고 이는 위에서 설명한 데로 모델이 complex하면 할 수록 더 심해진다고 말 할 수 있습니다. 이를 우리는 여러 설명모델 혹은 shpaley value로 설명 가능한 것입니다.<br>
   Q : 2. addictive feature attribtuion methods 3번 성질이 어떻게 유도가 되는지 궁급합니다<br>
   A : 저자에 의하면 Z'/i = 0으로 수렴합니다. 이를 이용하면 설명 모델의 attribution이 0으로 수렴함을 알 수 있고 이는 절대부등식을 만족시키게 됩니다. 따라서 특정 설명 모델이 가지고 있는 attrbition값은 그 전의 값보다는 작지 않다라는 결론을 나타낼 수 있습니다. 이러한 성질은 추후 shapley value를 도출해내 내는데 사용하게 됩니다.
